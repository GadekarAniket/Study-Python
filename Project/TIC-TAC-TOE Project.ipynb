# TIC-TAC-TOE Project

```python
#creating the Tic-tac-toe game below using dictionary, funcions, loop, condition statement.

TTT_Board = {'1':' ', '2':' ', '3':' ',
             '4':' ', '5':' ', '6':' ',
             '7':' ', '8':' ', '9':' '}

board_keys = []
            
for key in TTT_Board:
    board_keys.append(key)

def executeboard(board):
    print(board['1'] + '  ||' + board['2'] + '  ||' + board['3'] )
    print('===||===||===')
    print(board['4'] + '  ||' + board['5'] + '  ||' + board['6'] )
    print('===||===||===')
    print(board['7'] + '  ||' + board['8'] + '  ||' + board['9'] )
    
#selecting the turn of user.
def game():
    turn = 'x'
    count = 0
    
    for i in range(10):
        executeboard(TTT_Board)
        print("It's your turn," + turn + "  move to which place?")
        
        move = input()
        
        if TTT_Board[move] == ' ':
            TTT_Board[move] = turn
            count += 1
            
        else:
            print("That place is already filled.\nMove to which place?")
            continue
        
        
        #selecing the probability of winning conding, and set the result as per user acting weather it is win or tie.
        #winning conditions
        
        if count >= 5:
            if TTT_Board['1'] == TTT_Board['2'] == TTT_Board['3'] != ' ':   #Top row
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            elif TTT_Board['4'] == TTT_Board['5'] == TTT_Board['6'] != ' ':   #Middle row
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            elif TTT_Board['7'] == TTT_Board['8'] == TTT_Board['9'] != ' ':   #last row
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            elif TTT_Board['1'] == TTT_Board['4'] == TTT_Board['7'] != ' ':   #First coloum
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            elif TTT_Board['2'] == TTT_Board['5'] == TTT_Board['8'] != ' ':   #second coloum
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            elif TTT_Board['3'] == TTT_Board['6'] == TTT_Board['9'] != ' ':   #third coloum
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            elif TTT_Board['1'] == TTT_Board['5'] == TTT_Board['9'] != ' ':   #diagonal
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            elif TTT_Board['3'] == TTT_Board['5'] == TTT_Board['7'] != ' ':   #diagonal
                executeboard(TTT_Board)
                print("\n******Game Over******\n")
                print("<<<<<<" + turn + " Won >>>>>>")
                break
            
            #if no one win the game then it suppose to be declare as tie.
            
        if count == 9:
            print("\n******Game Over******\n")
            print("****** It's a Tie! ******")
                
            #now player have to change after every move.
            #i.e. after player_1 the next move has to be player_2.
            
        if turn == 'x':
            turn = '0'
        else:
            turn = 'x'
                
    #at last we have to ask the user if he want to play the game or not.               
    restart = input("Do you want to play again? (Y/N) :" )         
    if restart == "Y" or restart == "y":
        for key in board_keys:
            TTT_Board[key]= " "
                    
        game()

if __name__ == "__main__":
    game()
```
